```python
count = CountVectorizer()
count_matrix = count.fit_transform(df_subset['bag_of_words'])
```

```python
indices = pd.Series(df_subset.index)
indices[:5]
```

```python
cosine_sim = cosine_similarity(count_matrix, count_matrix)
cosine_sim
```

```python
def recco(title, cosine_sim = cosine_sim):
    reco_prods = []
    idx = indices[indices == title].index[0]
    score_series = pd.Series(cosine_sim[idx]).sort_values(ascending = False)
    top_10_indexes = list(score_series.iloc[1:30].index)
    for i in top_10_indexes:
        reco_prods.append(list(df_subset.index)[i])
    return reco_prods    
```
